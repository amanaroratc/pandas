## Filtering Data

```python
new_df=df.loc[(df['Type 1']=='Grass') & (df['Type 2']=='Poison') & (df['HP']>70)]
new_df.to_csv('filtered.csv')
new_df.head()
#observe that we got old indices
```

```python
#RESET INDEX
new_df.reset_index(drop=True, inplace=True)
new_df
```

```python
#the names 'Mega' is a little annoying?

df_nomega=df.loc[~df['Name'].str.contains('Mega')]
#very powerful function, we can remove all the names that had 'Mega' in them
df_nomega
```

```python
df.loc[df['Type 1'].str.contains('fire|grass', flags=re.I, regex=True)]
#why regex = True ???? regular expression. Check documentaion
#had to import re
#the flag is to ignorecase
```

What if I want to get pokemons whos name start with 'Pi'?

```python
df.loc[df['Name'].str.contains('pi[a-z]*', flags=re.I, regex=True)]
#notice we have everything containing 'pi'
#not what we wanted, eh?
```

```python
df.loc[df['Name'].str.contains('^pi[a-z]*', flags=re.I, regex=True)]
#THERE WE GO! :)
```

## Conditional Changes

```python
#what if I like 'Flamer' instead of 'Fire'?
df.loc[df['Type 1']=='Fire',  'Type 1']='Flamer'

#print(df)
#lets change it back, btw
df.loc[df['Type 1']=='Flamer',  'Type 1']='Fire'

df
```

```python
#can also change multiple parameters at a time
df.loc[df['Total']>500, ['Generation', 'Legendary']]=['TEST 1', 'TEST 2']
df
```

## Aggregate Statistics (Groubpy)

```python
df.groupby(['Type 1']).mean().sort_values('Defense', ascending=False)

#WOOHOO, steel and rock pokemons have more avg defense!
```

```python
df.groupby(['Type 1']).mean().sort_values('Total', ascending=False)
#overall, dragons are quite powerful :D
```

```python
df['count']=1

df.groupby(['Type 1']).count()['count']
```

```python
df.groupby(['Type 1', 'Type 2']).count()['count']
```

## Working with large amounts of data

```python
new_df=pd.DataFrame(columns=df.columns)

# for df in pd.read_csv('modified.csv', chunksize=5):
#     print("CHUNK DF")
#     print(df)

for df in pd.read_csv('modified.csv', chunksize=5):
    results=df.groupby(['Type 1']).count()
    
    new_df=pd.concat([new_df, results])

new_df

#IDK THIS WAS WEIRD AT THE END OF VIDEO
```
